```python
import time
import zmq

context = zmq.Context()
socket = context.socket(zmq.REP)
socket.bind("tcp://*:5555")
```

```python
while True:
    #  Wait for next request from client
    message = socket.recv()
    print("Received request: %s" % message)

    #  Do some 'work'
    if msg == 'test':
        socket.send_string('done')
        print('reply done')
    else:
        print('reply nothing')
    
    #  Send reply back to client
    # socket.send(b"World")
```
